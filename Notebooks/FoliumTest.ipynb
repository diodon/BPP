## Adding raster to folium

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import xarray as xr
import folium
```

```python
nc = xr.open_dataset("../Data/test_DHWmax.nc")
```

```python
nc.DHW[0,:,:].plot()
```

```python
m = folium.Map([-20, 120], zoom_start=6, tiles='stamentoner')
m
```

```python
img = nc.DHW[0,:,:].values
bounds=[[-10., 110.], [-30., 130.]]
folium.raster_layers.ImageOverlay(img, 
                                  bounds=bounds,
                                  opacity=0.5,
                                  colormap=cm.viridis
                                  ).add_to(m)
```
