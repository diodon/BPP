## Adding raster to folium

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import xarray as xr
import folium
```

```python
nc = xr.open_dataset("../../Data/test_DHWmax.nc")
```

```python
nc.DHW[0,:,:].plot()
```

```python
m = folium.Map([-20, 120], zoom_start=3, tiles='stamentoner')
m
```

```python
img = nc.DHW[0,:,:].values
bounds=[[-35., -180.],[35., 180.]]

m = folium.Map([-20, 120], zoom_start=3, tiles='stamentoner')
folium.raster_layers.ImageOverlay(img, 
                                  bounds=bounds,
                                  origin='lower', 
                                  opacity=0.7,
                                  pixelated=False,
                                  colormap=cm.coolwarm
                                  ).add_to(m)

m
```

```python
img2 = "../../Data/test_DHWmax_1985.tif"
bounds=[[-35., -180.],[35., 180.]]

m2 = folium.Map([-20, 120], zoom_start=3, tiles='stamentoner')
folium.raster_layers.ImageOverlay(img2, 
                                  bounds=bounds,
                                  origin='lower', 
                                  opacity=0.7,
                                  colormap=cm.coolwarm
                                  ).add_to(m2)

m
```
